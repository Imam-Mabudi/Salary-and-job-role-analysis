# DATA VISUALIZATION AND EXPLORATORY ANALYSIS

```python
data['Age'].value_counts()
```

```python
otal = len(data)
plt.figure(figsize = (35, 15))
plt.title('AGE')
ax = sns.countplot(x = 'Age', data = data)
for p in ax.patches:
    percentage = '{0:.0f}%'.format(p.get_height() / total * 100)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height() + 5
    ax.annotate(percentage, (x, y), ha = 'center')
plt.show()
```

```python
#for gender

data['Gender'].value_counts().plot(kind='pie')
plt.title('Gender(Pie Chart)', size = 25, color = 'blue')
plt.figure(figsize = (20,10))
```

```python
plt.figure(figsize = (8,5))

sns.countplot(data['Gender'])

plt.title('Gender Value Counts')
```

```python
# Countplot with percentage for gender

total = len(data)
plt.figure(figsize = (10, 6))
plt.title('Gender')
ax = sns.countplot(x = 'Gender', data = data)
for p in ax.patches:
    percentage = '{0:.0f}%'.format(p.get_height() / total * 100)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height() + 5
    ax.annotate(percentage, (x, y), ha = 'center')
plt.show()
```

##### Male are 60% populated while female are 40%

```python
#for Job role

data['Job Role'].value_counts().plot(kind='pie')
plt.title('Job Role(Pie Chart)', size = 35, color = 'blue')
plt.figure(figsize = (30,20))
```

```python
total = len(data)
plt.figure(figsize = (18, 6))
plt.title('Job Role')
ax = sns.countplot(x = 'Job Role', data = data)
for p in ax.patches:
    percentage = '{0:.0f}%'.format(p.get_height() / total * 100)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height() + 5
    ax.annotate(percentage, (x, y), ha = 'center')
plt.show()
```

### The business developer role have more staff, followed by Engineer and Technician

```python
otal = len(data)
plt.figure(figsize = (28, 8))
plt.title('Working Years')
ax = sns.countplot(x = 'Total Working Years', data = data)
for p in ax.patches:
    percentage = '{0:.0f}%'.format(p.get_height() / total * 100)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height() + 5
    ax.annotate(percentage, (x, y), ha = 'center')
plt.show()
```

### 10 years is the longest working years, followed by 6 years 

```python
# Exploring the distributions and the relationships between the job role vs the years in common role

plt.figure(figsize = (18, 10))
sns.barplot(x = 'Job Role', y = 'Years In Current Role', data = data)
```

#### The higest years in current job role is Manager followed by RnD Director
